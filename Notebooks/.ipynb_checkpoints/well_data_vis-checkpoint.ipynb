# Data Read In

```python
data_dir = "C:/Users/jinsu.elhance/Box/Wells/WellsDatasheets"
```

```python
full_well_data = pd.read_csv(f"{data_dir}/HistoricalWellSynthesis/Data/full_well_data.csv")
full_well_data['datetime'] = pd.to_datetime(full_well_data['datetime'])
```

# Assessing Data Quality

```python
#Looking for duplicate timestamps per well.
full_well_data.groupby(["well", "datetime"]).agg("COUNT")
```

```python
sns.catplot(full_well_data, x="datetime", y="well");
```

```python
for well in full_well_data['well'].unique():
    well_data = full_well_data.loc[full_well_data['well'] == well]
    sns.lineplot(well_data, x='datetime', y='level_corrected')
    plt.show()
```

```python
levels = sns.FacetGrid(full_well_data, col="well")
levels.map(sns.scatterplot, "datetime", "level_corrected")
```

```python
well_data = full_well_data.loc[full_well_data['well'] == "oaks5"][['datetime', 'level_corrected']].dropna()
sns.lineplot(well_data, x='datetime', y='level_corrected')
plt.show()
```
